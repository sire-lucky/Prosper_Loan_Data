### Converting IncomeRange into ordered categorical types

```python
ordinal_var_dict = {'IncomeRange' : ['Not displayed', 'Not employed', '$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999',
                                   '$75,000-99,999','$100,000+']}

for i in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = ordinal_var_dict[i])
    loan_subset1[i] = loan_subset1[i].astype(ordered_var)
```

### what is the monthly income distribution?

```python
#Creating a Histogram to show the distribution of the Stated Monthly Income
plt.figure(figsize = [8, 5])
bin = np.arange(0,  loan_subset1['StatedMonthlyIncome'].max()+200, 200)
plt.hist(data =  loan_subset1, x = 'StatedMonthlyIncome', bins = bin)
plt.xlim(0, 20000);
plt.xlabel('Monthly Income ($)');
plt.ylabel('Frequency');
plt.title('Monthly Income Distribution');
```

### creating def function for label to avoid repetitive coding

```python
def label(x, y, z):
    """
    Args:
    x(str): x-axis title
    y(str): y-axis title
    z(str): main title
    
    Returns:
    None
    """
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(z)
    plt.show()
```

### What month is loan most issued?
#Creating a barchart to show Month distribution

```python
def loan_month_plot():
    #Convert LoanOriginationDate to Datetime
    loan_subset1['LoanOriginationDate'] = pd.to_datetime(loan_subset1.LoanOriginationDate)
    #Extracting month from the 'LoanOriginationDate' variable
    loan_subset1['Month'] = loan_subset1['LoanOriginationDate'].dt.month
    #Plotting
    plt.figure(figsize = [8, 5])
    base_color = sb.color_palette()[0]
    sb.countplot(data = loan_subset1, x = 'Month', color = base_color)
    label('Month', 'Frequency', 'Month Distribution');
       
loan_month_plot()
```

month with the highest number of borrowers is January

### What year has the highest number of borrowers?

```python
#A bar chart showing the distribution of years
def loan_year_plot():
    #Convert LoanOriginationDate to Datetime and extracting year
    loan_subset['Year'] = pd.DatetimeIndex(loan_subset['LoanOriginationDate']).year
    
    plt.figure(figsize = [8, 5])
    base_color = sb.color_palette()[0]
    
    #To convert Year into ordinal variable
    year_order = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
    sb.countplot(data = loan_subset, x = 'Year', color = base_color, order = year_order)
    label('Year', 'Frequency', 'Year Distribution');
        
loan_year_plot()
```

2013 is the year with the highest number of borrowers

### Using Subplot to investigate the relationship between two categoriacal variables

```python
#Converting boolean to string
loan_subset1['IsBorrowerHomeowner'] = loan_subset['IsBorrowerHomeowner'].map({True: 'Yes', False: 'No'})

#Subplot 1: EmploymentStatus vs. IsBorrowerHomeowner 
plt.figure(figsize = [12, 8])
ax = plt.subplot(3, 1, 1)
sb.countplot(data = loan_subset, x = 'EmploymentStatus', hue = 'IsBorrowerHomeowner', palette = 'Blues')

#Subplot 2: IncomeRange vs. IsBorrowerHomeowner 
ax = plt.subplot(3, 1, 2)
sb.countplot(data = loan_subset, x = 'IncomeRange', hue = 'IsBorrowerHomeowner', palette = 'Blues')
plt.xticks(rotation = 30)
ax.legend(ncol = 2)
```

The majority of the borrowers have an IncomeRange between 25k - 49.999k, and they do not own their homes.
